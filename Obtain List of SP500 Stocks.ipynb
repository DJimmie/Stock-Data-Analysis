```python
#Writing method 1

for i in os.listdir(path=Location):
    print (i)
    ticker=i.split('_')
    print(ticker[0])
    with open(Location_1,'a',newline='') as ticks:
        daywriter=csv.writer(ticks)
        daywriter.writerow([ticker[0]])
    ticks.close
        
```

```python
#Writing method 2

with open(Location_1,'w',newline='') as ticks:
    daywriter=csv.writer(ticks)
    
    for i in os.listdir(path=Location):
        #print (i)
        ticker=i.split('_')
        print(ticker[0])
        
        daywriter.writerow([ticker[0]])
ticks.close
```

```python
#Reading
my_stocks=[]
with open(Location_1,'r') as get_tickers:
    dayreader=csv.reader(get_tickers)
    print (dayreader)
    for the_symbols in dayreader:
        #print (the_symbols)
        my_stocks.append(the_symbols[0])
        
get_tickers.close
```
